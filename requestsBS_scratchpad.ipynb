```python
url = 'https://www.python.org/~guido/'
```

```python
type(response)
```

```python
html_doc = response.text
```

```python
print(type(html_doc))
```

```python
soup = BeautifulSoup(html_doc)
```

```python
print(soup)
```

```python
pretty_soup = soup.prettify()
```
