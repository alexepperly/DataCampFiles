```python
# cumulative distribution function example from ChatGPT:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

test_scores = np.random.randint(0, 101, size=100)

test_scores
```

```python
plt.hist(test_scores, bins=10, range=(0, 100), density=True, cumulative=True)
plt.xlabel('Test Scores')
plt.ylabel('Probability')
plt.title('CDF of Test Scores')
plt.show()
```

```python
die = pd.Series([1, 2, 3, 4, 5, 6])
samp_5 = die.sample(5, replace=True)
print(samp_5)
```

```python
np.mean(samp_5)
```

```python
roll_meanz = []
for i in range(1000):
    samp_5 = die.sample(5, replace=True)
    roll_meanz.append(np.mean(samp_5))

plt.hist(roll_meanz, bins=7);
    
```
