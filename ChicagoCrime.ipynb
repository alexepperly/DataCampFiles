```python
crime_counts = chi_df['Primary Type'].value_counts()
```

```python
crime_counts
```

```python
plt.figure(figsize=(12, 6))  # Optional: Adjust the figure size

crime_counts.plot(kind='bar')
plt.title('Distribution of Crimes in Chicago')
plt.xlabel('Crime Type')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Optional: Rotate x-axis labels for better readability
plt.tight_layout()       # Optional: Adjust layout for better spacing
plt.show()
```

```python
# Generating a list of unique colors for each bar
num_colors = len(crime_counts)
colors = plt.cm.tab20.colors[:num_colors]

plt.figure(figsize=(12, 6))
bars = plt.bar(crime_counts.index, crime_counts.values, color=colors)
plt.title('Distribution of Crimes in Chicago')
plt.xlabel('Crime Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()

# Adding legend with proper crime type labels
plt.legend(bars, crime_counts.index)

plt.show()
```

```python
# Generating a list of unique colors for each bar
num_colors = len(crime_counts)
colors = sns.color_palette('Set3', n_colors=num_colors)

plt.figure(figsize=(12, 6))
sns.barplot(x=crime_counts.index, y=crime_counts.values, palette=colors)
plt.title('Distribution of Crimes in Chicago')
plt.xlabel('Crime Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()

# Adding legend with proper crime type labels
legend_labels = [crime.replace('\n', ' ') for crime in crime_counts.index]  # Handling potential newline characters

plt.show()
```
